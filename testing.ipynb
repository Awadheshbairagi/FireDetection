```python
import os
import cv2
from PIL import Image
from IPython.display import display
```

```python
# Get the current working directory
cwd = os.getcwd()

# Get the parent directory
parent_dir = os.path.dirname(cwd)

# Change the CWD to the `outputs` folder in the parent directory
img_dir = os.path.join(parent_dir, "outputs")

# Get the list of all subdirectories in the current directory
subdirs = os.listdir(img_dir)

# Get the last subdirectory in the list
last_subdir = subdirs[-1]

# Get the path to the last subdirectory
last_subdir_path = os.path.join(img_dir, last_subdir)

# Get the list of all files in the last subdirectory
files = os.listdir(last_subdir_path)

# Filter the list of files to only include image files
image_files = []
for file in files:
    if file.endswith(".jpg") or file.endswith(".jpg"):
        image_files.append(file)

# Display all image files in the last subdirectory
for image_file in image_files:
    img = Image.open(os.path.join(last_subdir_path, image_file))
    # Display the image
    display(img)
```

```python
def get_img_path():
    cwd = os.getcwd()
    parent_dir = os.path.dirname(cwd)
    img_dir = os.path.join(parent_dir, "outputs")
    dirs = os.listdir(img_dir)
    last_dir = dirs[-1]
    req_dir = os.path.join(img_dir, last_dir)
    
    return req_dir

dir = get_img_path()
print(dir)

for f in os.listdir(dir):
    img_path = os.path.join(dir, f)
    img = Image.open(img_path)
    display(img)
```
